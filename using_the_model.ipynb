```python
import glob
import re
import os
from transformers import pipeline
model=pipeline("question-answering",model="Aharneish/qa-model") #link to my model this can be changed to the model you have trained
```

```python
text = glob.glob('/content/*.txt')
con=[]
def clear(dirname):
    for file in glob.glob(dirname+'*'):
        os.remove(file)

def clean(text):
    text = re.sub(r' +', ' ', text)
    return text
for i in text:
    with open(i, 'r') as f:
        cur_text = f.readlines()
        f.close()
    text = clean(cur_text[0])
    con.append(cur_text)
context = ''.join(str(v) for v in con)
```

```python
len(context)
```

```python
q=input('enter the question you wan to ask : ')
ans=model(question=q,context=context)
print('according to thirukkural the answer is' ,ans['answer']) # prediction
```
